    - `jajang_price.csv`

    - Í≤∞Ï∏°ÏπòÎäî Ìï¥Îãπ ÌñâÏùÑ ÏÇ≠Ï†úÌïòÏãúÏò§. (`dropna()`)
    - Ïò§Î∏åÏ†ùÌä∏ ÌÉÄÏûÖ Ïª¨ÎüºÏóê ÎåÄÌï¥ÏÑúÎäî Ïõê-Ìï´ Ïù∏ÏΩîÎî©(`get_dummies`)ÏùÑ Ï†ÅÏö©ÌïòÏó¨ ÏàòÏπòÌòïÏúºÎ°ú Î≥ÄÌôò

````python
# Ï†ïÎãµ ÏΩîÎìú
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞ Î∞è Ï†ÑÏ≤òÎ¶¨
df = pd.read_csv('jajang_price.csv')
df.dropna(inplace=True)
df = pd.get_dummies(df, drop_first=True)

# X, y Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
X = df.drop('price', axis=1)
y = df['price']

# ÌõàÎ†®/Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï†
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# # Ïä§ÏºÄÏùºÎßÅ
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

````

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import tensorflow as tf

# ÎûúÎç§ ÏãúÎìú Í≥†Ï†ï
tf.random.set_seed(100)


model = Sequential()
model.add(Dense(256, activation='relu', input_shape=[X_train_scaled.shape[1]]))
model.add(Dropout(0.3))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))  # ÌÅ¥ÎûòÏä§ ÏàòÎßåÌÅº ÎÖ∏Îìú


## ÌïôÏäµ Î∞©Î≤ï ÏÑ§Ï†ï
model.compile(loss='mse',optimizer='adam',metrics=['mse'])#------->ÌöåÍ∑Ä


# EarlyStopping ÏΩúÎ∞± ÏÑ§Ï†ï
early_stop = EarlyStopping(monitor='val_loss', patience=10)

# ModelCheckpoint ÏΩúÎ∞± ÏÑ§Ï†ï
model_ckp = ModelCheckpoint(
    filepath='best_jajang_model.keras',
    # filepath='best_jajang_model.h5',
    monitor='val_loss',
    save_best_only=True,
    # verbose=1
)

hist = model.fit(
        X_train_scaled, y_train,
        epochs=200, #Ïñ¥Ï∞®Ìîº Ï°∞Í∏∞ Ï¢ÖÎ£åÍ∞Ä ÏûàÏúºÎãà Ï∂©Î∂ÑÌûà ÌÅ∞ Ïà´Ïûê ÏÑ§Ï†ï
        batch_size=32,
        validation_data=(X_test_scaled, y_test),
        callbacks=[early_stop, model_ckp],
)

ÎãµÏïà02= 'Ï°∞Í∏∞Ï¢ÖÎ£åÏ†ÅÏö©' if len(hist.history['loss']) < 200   else 'Ï°∞Í∏∞Ï¢ÖÎ£åÏ†ÅÏö©ÏïàÎê®'
ÎãµÏïà02
````
