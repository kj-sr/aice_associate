# **🚀 AICE Associate 최종 합격 전략**

## 1.  **순차 시험임을 기억하세요. ⛓️**
* 앞선 코드의 결과가 다음 문제에 영향을 줍니다.
* 단계마다 `df.head()`, `df.shape`, `df.info()`등으로 중간 결과를 꼭 확인하세요.

## 2.  **EDA와 전처리에 많은 공부시간을 투자하세요. ⏳**
* 합격의 6~70%가 데이터 탐색과 전처리에서 결정됩니다.
* 이 구간에서 안정적인 점수를 확보하는 것이 중요합니다.

## 3.  **오픈북보다 빠른 `?`를 활용하세요. ⚡**
* 함수나 파라미터가 헷갈릴 때 가장 빠른 해결책입니다.
* (예: `df.fillna?`, `train_test_split?`)

## 4.  **'날코딩'으로 실전 감각을 기르세요. ⛔**
* 시험 환경엔 코드 자동 완성, 복사/붙여넣기 기능이 없습니다.
* 평소에 자동 완성 기능을 끄고 연습해야 당황하지 않습니다.




## 5.  **`inplace=True`의 함정을 피하세요. ✍️**
* `df.dropna(inplace=True)` 대신 `df = df.dropna()`를 사용을 추천합니다.
* `inplace=True`는 데이터를 통째로 날릴 수 있습니다.
* 지시사항에 명시적으로 사용하라고 할 때만 사용합니다.

## 6.  **패턴 공략과 안전한 테스트는 필수입니다. 💡**
* 문제 유형별 코드 패턴을 익혀두세요.
* 테스트는 `df_test = df.copy()`로 원본을 보호하고, 제출 전 테스트 셀은 반드시 삭제하세요.
* 예상치 못한 상황을 대비해 수시로 **임시 저장**하는 습관을 들이세요.

## 7.  **AICE 샘플문항 풀어세요.**
* 시험 환경을 사전에 경험하세요.
* 샘플문항을 50분안에 풀 수 있도록 연습하세요.

# **📝 AICE Associate 시험 핵심 Q&A**

##**Q: 가장 효율적인 공부 순서는 어떻게 되나요?**
* AI 적용 프로세스대로 공부하는 것이 가장 효율적입니다.
* 특히, 점수 비중이 가장 높은 **데이터 전처리** 파트를 집중적으로 연습하는 것이 합격의 핵심입니다.

##**Q: 모델의 성능이 낮게 나와도 합격할 수 있나요?**
* 네, **합격할 수 있습니다.** Associate 등급은 모델의 성능 점수 자체보다, **주어진 요구사항에 맞춰 코드를 올바르게 작성하는 과정**을 평가합니다.
* 성능이 다소 낮더라도 지시사항(예: 결측치 처리, 스케일링 등)을 모두 수행했다면 충분히 합격 가능합니다.

##**Q: 코드를 다 외워야 하나요? 함수 이름이나 파라미터가 기억나지 않으면 어떡하죠?**
* 시간관계상 다 외울정도로 익숙해지는 것을 추천합니다.
* 다만, 기억나지 않을경우 **함수 뒤에 `?`를 붙여 실행**하세요.

##**Q: 머신러닝, 딥러닝에서 하이퍼파라미터가 너무 많은데, 이 의미를 모두 알아야 할까요?**
* **아니요, 모두 알 필요 없습니다.** Associate 등급에서는 대부분의 하이퍼파라미터를 **기본값(default)**으로 두고 문제를 풉니다.
* 세부적인 튜닝보다는, 모델을 생성하고 학습시키는 **기본적인 코드 구현**에만 집중하면 충분합니다.

##**Q: 알고리즘의 원리도 공부해야 할까요?**
* **깊은 수학적 원리까지는 필요 없습니다.**
* "회귀 문제는 LinearRegression", "분류 문제는 LogisticRegression, DecisionTree" 와 같이 **문제 유형에 맞는 적절한 모델을 선택하고, 구현할 수 있는 능력**만 있으면 됩니다.
