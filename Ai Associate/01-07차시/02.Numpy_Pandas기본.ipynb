# 2차시: Pandas는 NumPy 위에 만들어졌다

이 노트북에서는 `ndarray` 구조를 이해하고, Pandas가 NumPy를 내부적으로 사용하고 있음을 실습으로 확인합니다.

### 판다스 데이터 구조
#### 1. 시리즈 (Series)
- 1차원 데이터 구조
- 구성: `index` + `value`
- 예: 하나의 열(column)만 있는 데이터

#### 2. 데이터프레임 (DataFrame)
- 2차원 테이블 형태의 데이터 구조
- 구성: `index`, `columns`, `values`
- 엑셀과 유사한 행-열 구조

#### 3. 구성 요소 설명
- `index`: 행 번호 (예: 0, 1, 2, ...)
- `columns`: 열 이름 (예: 'mango', 'apple', 'banana')
- `values`: 셀 안에 들어 있는 실제 데이터 값



## 1. NumPy 배열 생성 및 속성 확인

```python
import numpy as np # NumPy 라이브러리 임포트
```

```python
# 2행 2열의 배열 생성
alist = [[90, 85], [78, 92]] #파이썬의 리스트
arr = np.array(alist)   #NumPy의 배열
print("NumPy 배열:")
print(arr)
```

```python
# 배열 정보 확인
print("자료형:", type(arr))
print("차원 수:", arr.ndim)
print("배열 형태:", arr.shape)
print("데이터 타입:", arr.dtype)
```

```python
# ndarray에서 값 추출
print("첫 번째 행:", arr[0])  
print("두 번째 열:", arr[:, 1])
```

## 2. Pandas의 시리즈(Series) 이해하기
* index와 values로 이루어진 데이터구조이다.

```python
import pandas as pd  # 라이브러리 임포트
```

```python
sd = pd.Series([4,5,6,3,1])
sd
```

```python
#타입체크
type(sd)
```

```python
# 값만 추출하여 ndarray로 변환
sd.values
```

```python
#타입체크
type(sd.values)
```

```python
#값 추출
type(sd.values[2])
if sd.values[2]==2:
    print('hello')
```

## 3. Pandas의 DataFrame 이해하기

```python
df = pd.DataFrame({
    "Mango": [4,5,6,3,1],
    "Apple": [5, 4, 3, 0, 2],
    "Banana": [2,3,5,2,7]    
})
df
```

```python
#타입 체크
type(df)
```

```python
#컬럼 리스트
df.columns
```

```python
#인덱스 리스트
df.index
```

```python
#값 리스트
df.values
```

```python
#값 타입 체크
type(df.columns)
type(df.index)
type(df.values)
```

## 4. 컬럼 이름 변환

```python
# 컬럼 이름 변경
df.columns = ['mango', 'apple', 'banana']
df
```

## 5. 인덱스 초기화 (reset_index)

```python
# `inplace=True`를 쓰면 원본 df에 바로 적용
df.set_index('mango',inplace=True)
df
```

```python
df.reset_index(inplace=True)  # 인덱스 초기화
df
```
