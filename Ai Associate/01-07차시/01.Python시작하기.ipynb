# 1차시: Python 기초

## 1. 변수와 기본 자료형

```python
# 변수 선언
x = 10
y = 3.14
name = "Tom"
is_passed = True
```

```python
# 변수의 자료형 확인
print(type(x))      # int
print(type(y))      # float
print(type(name))   # str
print(type(is_passed))  # bool
```

## 2. 리스트

```python
# 리스트 선언
scores = [95, 88, 76, 53, 90]
print(scores)

# 인덱싱
print(scores[0])  # 첫 번째 점수
print(scores[1:4]) # 두 번째부터 네 번째 점수
```

## 3. 딕셔너리

```python
# 딕셔너리 선언
student = {"name": "Tom", "score": 95}

# 값 접근
print(student["name"])
print(student["score"])
```

## 4. 조건문

```python
# 조건문 사용 예시
if student["score"] >= 90:
    print("우수 학생입니다.")
else:
    print("노력이 필요합니다.")
```

## 5.반복문

```python
for element in scores:
    print(element)
```
